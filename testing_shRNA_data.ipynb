```python
# Read in signature info
sig_info = pd.read_csv(siginfopath, sep="\t")

# Get the signatures associated with gene knockdown via shRNA
sh_rna_sigs = sig_info['sig_id'][sig_info['pert_type'] == 'trt_sh'] # TODO: Or t
num_sigs = len(sh_rna_sigs)
print('num', num_sigs)
jump_size = 10000
i = 0
j = jump_size
sh_rna_sigs_subset = sh_rna_sigs[i:j]
len(sh_rna_sigs_subset)
```

```python
# Get the data for the gene signatures
# takes about 4 min for 10k sigs
import datetime
print('before', datetime.datetime.now())
sh_rna_gctoo = parse(filepath, cid=sh_rna_sigs_subset, convert_neg_666=True)
print('after', datetime.datetime.now())
```

```python
gene_ids = pd.read_csv(geneinfopath, sep="\t")
print(gene_ids.shape)
gene_ids
```

```python
sh_rnas = sig_info[sig_info['pert_type'] == 'trt_sh'].pert_iname.unique()
genes = gene_ids.pr_gene_symbol.unique()
has_gene_info, no_gene_info = [], []
for sh_rna in sh_rnas:
    if sh_rna in genes:
        has_gene_info.append(sh_rna)
    else:
        no_gene_info.append(sh_rna)
is_landmark = []
no_landmark = []
for sh_rna in has_gene_info:
    if sh_rna in gene_ids[gene_ids['pr_is_lm'] == 1].pr_gene_symbol.unique():
        is_landmark.append(sh_rna)
    else:
        no_landmark.append(sh_rna)
```

```python
print('has', len(has_gene_info))
print('no', len(no_gene_info))
print('landmark', len(is_landmark))
print('no landmark', len(no_landmark))
```
